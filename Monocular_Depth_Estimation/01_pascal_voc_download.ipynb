```python
from google.colab import drive
import torchvision
from torchvision import transforms

# Google Drive 마운트
drive.mount('/content/drive')

# 데이터 변환 정의 (필요에 따라 조정 가능)
transform = transforms.Compose([
    transforms.ToTensor()
])

# Google Drive의 data 폴더 경로 설정
drive_data_path = '/content/drive/My Drive/data'

# VOC 데이터셋 로드 (훈련 데이터)
voc_train = torchvision.datasets.VOCDetection(root=drive_data_path, year='2012', image_set='train', download=True, transform=transform)
```

```python
# 전체 이미지 개수 확인
num_images = len(voc_train)
print(f"전체 이미지 개수: {num_images}")

# 데이터셋의 예시 확인
print(voc_train[0])
```

```python
from google.colab import drive
import torchvision
from torchvision import transforms
from collections import defaultdict
from PIL import Image
# Google Drive 마운트
drive.mount('/content/drive')

# Google Drive의 data 폴더 경로 설정
drive_data_path = '/content/drive/My Drive/data'

# 각 클래스마다 이미지를 저장할 딕셔너리 초기화
class_to_images = defaultdict(list)

# 데이터셋을 순회하면서 각 이미지의 클래스를 확인하고 저장
for img_idx in range(len(voc_train)):
    image, target = voc_train[img_idx]
    for obj in target['annotation']['object']:
        class_name = obj['name']
        if len(class_to_images[class_name]) < 5:
            class_to_images[class_name].append((image, target))
        # 이미 5개의 이미지가 저장된 경우 건너뛰기
        if all(len(imgs) >= 5 for imgs in class_to_images.values()):
            break

# 각 클래스별로 상위 5개의 이미지 확인
for class_name, images in class_to_images.items():
    print(f"Class: {class_name}")
    for img_idx, (image, target) in enumerate(images):
        print(f"  Image {img_idx + 1}: {target['annotation']['filename']}")
```

```python
import os
save_path = '/content/drive/My Drive/data/voc_top5'
if not os.path.exists(save_path):
    os.makedirs(save_path)

for class_name, images in class_to_images.items():
    class_save_path = os.path.join(save_path, class_name)
    if not os.path.exists(class_save_path):
        os.makedirs(class_save_path)
    print(f"Class: {class_name}")
    for img_idx, (image, target) in enumerate(images):
        img_filename = f"{class_name}_{img_idx + 1}.jpg"
        img_save_path = os.path.join(class_save_path, img_filename)
        image_pil = Image.fromarray((image.numpy().transpose(1, 2, 0) * 255).astype('uint8'))
        image_pil.save(img_save_path)

        annot_filename = f"{class_name}_{img_idx + 1}.txt"
        annot_save_path = os.path.join(class_save_path, annot_filename)
        with open(annot_save_path, 'w') as f:
            f.write(str(target))

        print(f"  Image {img_idx + 1}: {target['annotation']['filename']} saved as {img_filename} and {annot_filename}")
```
